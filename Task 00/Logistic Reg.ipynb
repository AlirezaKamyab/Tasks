# Logistic Regression

```python
import numpy as np
import pandas as pd
from sklearn.linear_model import LogisticRegression
from sklearn.preprocessing import StandardScaler, OneHotEncoder, OrdinalEncoder
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('data_task003.csv')
```

```python
df.head()
```

```python
department_ = np.array(df['Department']).reshape(-1, 1)
ohe = OneHotEncoder(drop='first').fit(department_)
department_dummy = ohe.transform(department_).toarray()
department_df = pd.DataFrame(department_dummy, columns=ohe.categories_[0][1:])
department_df
```

```python
salaries = np.array(df['salary']).reshape(-1, 1)
ode = OrdinalEncoder(categories=[['low', 'medium', 'high']])
salaries = ode.fit_transform(salaries)
salary_df = pd.DataFrame(salaries, columns=['Salary'])
```

```python
df = pd.concat([df.drop(['Department', 'salary'], axis=1), department_df, salary_df], axis=1)
df.head()
```

```python
X = np.array(df.drop(['Salary'], axis=1))
y = np.array(df['Salary'])
```

```python
X
```

```python
y
```

```python
scalar = StandardScaler().fit(X)
X = scalar.transform(X)
```

```python
x_train, x_test, y_train, y_test = train_test_split(X, y, train_size=0.8, shuffle=True)
logistic_reg = LogisticRegression().fit(x_train, y_train)
logistic_reg.score(x_test, y_test)
```
