# Logistic Regression

```python
import numpy as np
import pandas as pd
from sklearn.linear_model import LogisticRegression
from sklearn.preprocessing import StandardScaler, OneHotEncoder, OrdinalEncoder
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('data_task003.csv')
```

```python
df.head()
```

```python
department_ = np.array(df['Department']).reshape(-1, 1)
ohe = OneHotEncoder(drop='first').fit(department_)
department_dummy = ohe.transform(department_).toarray()
department_df = pd.DataFrame(department_dummy, columns=ohe.categories_[0][1:])
department_df
```

```python
salaries = np.array(df['salary']).reshape(-1, 1)
ode = OrdinalEncoder(categories=[['low', 'medium', 'high']])
salaries = ode.fit_transform(salaries)
salary_df = pd.DataFrame(salaries, columns=['Salary'])
```

```python
df = pd.concat([df.drop(['Department', 'salary'], axis=1), department_df, salary_df], axis=1)
df.head()
```

```python
X = np.array(df.drop(['Salary'], axis=1))
y = np.array(df['Salary'])
```

```python
X
```

```python
y
```

```python
scalar = StandardScaler().fit(X)
X = scalar.transform(X)
```

```python
x_train, x_test, y_train, y_test = train_test_split(X, y, train_size=0.8, shuffle=True)
logistic_reg = LogisticRegression().fit(x_train, y_train)
logistic_reg.score(x_test, y_test)
```

```python
logistic_reg.coef_
```

```python
prediction_result = logistic_reg.predict(x_test)
```

```python
x_test = scalar.inverse_transform(x_test)
```

```python
result_df = pd.DataFrame(x_test[:, 0:8], columns=df.columns[0:8])
result_df.head(5)
```

```python
result_department = ohe.inverse_transform(x_test[:, 8:17])
result_salary = ode.inverse_transform(prediction_result.reshape(-1, 1))
```

```python
result_df['Department'] = pd.Series(result_department.reshape(-1,))
result_df['Salary'] = pd.Series(result_salary.reshape(-1,))
result_df['left'] = result_df['left'].apply(lambda x: 0 if x < 0.5 else 1)
result_df['promotion_last_5years'] = result_df['promotion_last_5years'].apply(lambda x: 0 if x < 0.5 else 1)
result_df.head()
```

```python
result_df.to_csv('Logistic_reg.csv', index=False)
```
