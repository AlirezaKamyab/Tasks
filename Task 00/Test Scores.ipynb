```python
from sklearn.linear_model import SGDRegressor
from sklearn.preprocessing import StandardScaler
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import math
```

```python
df = pd.read_csv("test scores.csv", skiprows=1)
df
```

```python
X = np.array(df['math'], dtype='int64')
y = np.array(df['cs'], dtype='int64')
X = X.reshape((-1, 1))
```

```python
plt.figure()
plt.scatter(X, y)
ax = plt.gca()
ax.set_title('cs / math plot')
ax.set_xlabel('Math')
ax.set_ylabel('CS');
```

```python
scale = StandardScaler()
scale = scale.fit(X)
reg = SGDRegressor(alpha=0.001, verbose=0)
```

```python
def cost(X, y, regressor:SGDRegressor):
    m, _ = X.shape
    err = (regressor.predict(X) - y) ** 2
    cost_sum = err.sum() / (2 * m)
    return cost_sum
```

```python
def train(X, y, regressor:SGDRegressor):
    prev_cost = 0
    while(True):
        regressor.partial_fit(X, y)
        c = cost(X, y, regressor)
        print(c)
        if np.isclose(c, prev_cost, rtol=1e-20): break
        prev_cost = c
```

```python
train(scale.transform(X), y, reg)
```

```python
reg.score(scale.transform(X), y)
```

```python
coef = reg.coef_
bias = reg.intercept_
print(f"coef={coef} bias={bias}")
```

```python
plt.figure()
plt.scatter(X, y)
ax = plt.gca()
ax.set_title('cs / math plot')
ax.set_xlabel('Math')
ax.set_ylabel('CS');
test_x = np.arange(X.min(), X.max() + 1).reshape(-1, 1)
test_y = scale.transform(test_x) * coef + bias
plt.plot(test_x, test_y)
```
