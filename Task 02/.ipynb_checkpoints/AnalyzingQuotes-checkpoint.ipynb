# Analyzing Quotes

```python
import pandas as pd
import matplotlib.pyplot as plt
from wordcloud import WordCloud
import spacy
import seaborn as sns

nlp = spacy.load('en_core_web_sm')
```

```python
df = pd.read_csv('quotes.csv')
df.head()
```

```python
all_texts = ''

for text in df.text:
    all_texts += str(text) + '\n'
print('Done')
```

```python
def token_frequency(txt, not_include=[], only_include=[], verbos=False):
    if len(not_include) != 0 and len(only_include) != 0: 
        raise Exception('Cannot specify both "not_include" and "only_include"')
    words = dict()
    doc = nlp(txt)
    for token in doc:
        if token.pos_ in not_include: continue
        if len(only_include) != 0 and token.pos_ not in only_include: continue
        if nlp.vocab[token.lemma_].is_stop: continue
        if token.lemma_ not in words.keys():
            words[token.lemma_] = 1
        else: words[token.lemma_] += 1
        if verbos: print(f"{token.text:{16}} {token.pos_:{12}} {token.dep_:{12}} {token.lemma_:{16}} {spacy.explain(token.tag_)}")
    
    words = dict(sorted(words.items(), key=lambda item: item[1], reverse=True))
    return words
```

```python
words = token_frequency(all_texts, not_include=['PUNCT', 'SPACE', 'NUM'])
```

```python
words
```

```python
df_freq = pd.DataFrame({'word':words.keys(), 'frequency':words.values()})
df_freq.head()
```

```python
fig, ax = plt.subplots(5, 1, figsize=(20, 50))

for axes in ax:
    axes.set_ylim([0, df_freq.frequency.max()])

s = 0
e = 20
for axes in ax:
    sns.barplot(data=df_freq.iloc[s:e], x='word', y='frequency', ax=axes)
    s = e
    e += 20
```

```python
wordcloud = WordCloud(background_color='#FFFFFF', width=1200, height=800).generate_from_frequencies(words)
plt.figure(figsize=(20, 10))
plt.imshow(wordcloud)
```

As the topic is science, it is expected that the word "Science" will be the most frequently used word.
Also, "time" has been frequent in quotes; in my point of view "time" is an interesting subject which is not understood completely or at least difficult to do so; as a matter of fact the word "understand" is one of the mostly used words.
Religion, is always considered to be in contrast with science and that is why the word "religion" or perhaps "god" has been frequent.
Humen, looking at the skies, always wondered what is beyond and that also has been a topic of interest, this might have been a good reason that "Universe", "world", "earth" are one of the mostly used words.
Some other words are good to notice, like, "think", "understand", "knowledge", "fact", "truth", "know" . . . which could be related to science or scientific factsl
