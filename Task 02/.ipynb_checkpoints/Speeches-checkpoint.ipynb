<h1>Analyzing most famous speeches in history</h1>

```python
import spacy
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
from wordcloud import WordCloud

nlp = spacy.load('en_core_web_sm')
```

```python
speeches = []
for i in range(1, 4):
    with open(f'speech{i}.txt', 'r') as file:
        speeches.append(file.read())
        print(f'File {i} is completed!')
```

Function below is used to find the frequencies and will be used for each speech

```python
def token_frequency(txt, not_include=[], only_include=[], verbos=False):
    if len(not_include) != 0 and len(only_include) != 0: 
        raise Exception('Cannot specify both "not_include" and "only_include"')
    words = dict()
    doc = nlp(txt)
    for token in doc:
        if token.pos_ in not_include: continue
        if len(only_include) != 0 and token.pos_ not in only_include: continue
        if nlp.vocab[token.lemma_].is_stop: continue
        if token.lemma_ not in words.keys():
            words[token.lemma_] = 1
        else: words[token.lemma_] += 1
        if verbos: print(f"{token.text:{16}} {token.pos_:{12}} {token.dep_:{12}} {token.lemma_:{16}} {spacy.explain(token.tag_)}")
    
    words = dict(sorted(words.items(), key=lambda item: item[1], reverse=True))
    return words
```

<h2>Speech1</h2>
<h3>Theodore Roosevelt, “Duties of American Citizenship”</h3>

```python
print(speeches[0])
```

```python
speech1_freq_all = token_frequency(speeches[0], not_include=['PUNCT', 'SPACE', 'NUM'], verbos=False)
speech1_freq_verbs = token_frequency(speeches[0], only_include=['VERB'], verbos=False)
```

```python
df_speech1_all = pd.DataFrame({'word':speech1_freq_all.keys(), 'frequency':speech1_freq_all.values()})
df_speech1_verbs = pd.DataFrame({'word':speech1_freq_verbs.keys(), 'frequency':speech1_freq_verbs.values()})
```

```python
fig, ax = plt.subplots(2, 1, figsize=(20, 10))
sns.barplot(data=df_speech1_all.iloc[:20], x='word', y='frequency', ax=ax[0])
sns.barplot(data=df_speech1_verbs.iloc[:20], x='word', y='frequency', ax=ax[1])
ax[0].set_title('All')
ax[1].set_title('Verbs');
plt.subplots_adjust(top=1.2);
```

```python
wordcloud = WordCloud(background_color='#FFFFFF', width=1200, height=800)
speech1_wc_all = wordcloud.generate_from_frequencies(speech1_freq_all)

wordcloud = WordCloud(background_color='#FFFFFF', width=1200, height=800)
speech1_wc_verbs = wordcloud.generate_from_frequencies(speech1_freq_verbs)
```

```python
fig, ax = plt.subplots(2, 1, figsize=(20, 10))
ax[0].imshow(speech1_wc_all)
ax[0].set_title('All')

ax[1].imshow(speech1_wc_verbs)
ax[1].set_title('Verbs')
```
